```python
## add a city column
#olets use the .apply ( funnction)


def get_city(address):
    return address.split(',')[1]
    
all_data['City']=all_data['Purchase Address'].apply( lambda x: get_city(x))
all_data.head()
```

```python
#or this method
#all_data['Column']=all_data['Purchase Address'].apply( lambda x: x.split(',')[1])
#all_data.head()
```

```python
all_data=all_data.drop(columns='Column')
```

```python
def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ' )')
all_data.head()
```

## Question 3 what products are most often sold together

```python
all_data['Product'].value_counts()
```

```python
all_data['Product'].value_counts().plot()
```

```python
sns.countplot(all_data['Product'],orient='v')
```

```python
all_data
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]


df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df=df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count= Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
print(count)
```

```python
count.most_common(10)
```

```python
for key, value in count.most_common(10):
    print(key,value)
```
